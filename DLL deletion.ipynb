```python
class Node:
    def __init__(self,data):
        self.data = data
        self.pref = None
        self.nref = None

class DLL:
    def __init__(self):
        self.head = None
        
    def print__LL(self):
        if self.head is None:
            print("LL is empty")                       ##### forward traversal
        else:
            n = self.head
            while n is not None:
                print(n.data,"--->",end = " ")
                n = n.nref
    def print__LL_reverse(self):
        if self.head is None:
            print("LL is empty")
        else:
            n = self.head                               #####   backward traversal
            while n.nref is not None:
                n=n.nref
            while n is not None:
                print(n.data,"--->",end = " ")
                n= n.pref
    def insert_empty(self,data):
        if self.head is None:
            new_node = Node(data)
            self.head =new_node
        else:
            print("LL is not empty")
    
    
    def add_begin(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head  = new_node
        else:
            new_node.ref = self.head
            self.head.pref  = new_node
            self.head = new_node
            
    def add_end(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head  = new_node
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            n.nref = new_node
            new_node.pref = n

    def add_after(self,data,x):
        if self.head is None:     ##### this shows that LL is empty
            print("LL is empty")
        else:
            n  = self.head
            while n is not None:
                if x==n.data:
                    break
                n = n.nref
            if n is None:
                print("given node not present in DLL")
           
            else:
                new_node = Node(data)
                ####### for adding a new nide after a given node  in which nref is Not none
                new_node.nref = n.nref
                new_node.pref = n 
                if n.nref is not None:    ##### this means it it not the last node because in last node the n.nref is none 
                    n.nref.pref = new_node
                n.nref = new_node    
                                          
        
    def add_before(self,data,x):
         if self.head is None:     ##### this shows that LL is empty
            print("LL is empty")
         else:
            n  = self.head
            while n is not None:
                if x==n.data:
                    break
                n = n.nref
            if n is None:
                print("given node not present in DLL")
            
            else:
                new_node = Node(data)
                #### now if I want to add node before 2,3 or rest node but not first
                new_node.nref = n
                new_node.pref  = n.pref
                if n.pref is not None:
                    n.pref.nref = new_node
                else:
                    self.head = new_node
                n.pref = new_node   
                
                
                
                
                
                
    def delete_begin(self):
        if self.head is None:
            print("DLL is empty")
            return
        if self.head.nref is None:
            self.head = None
            print("DLL is empty after deleting the node")
            
        else:
            self.head  = self.head.nref
            self.head.pref = None
            
            
            
    def delete_end(self):
        
        if self.head is None:
            print("DLL is empty")
            return
        
        if self.head.nref is None:
            self.head = None
            print("DLL is empty after deleting nodes")
        
        else:
            n = self.head
            while n.nref is not None:
                  n= n.nref
            n.pref.nref = None
            
    
    def delete_by_value(self,x):
        if self.head is None:                   #### check DLL is empty or not
            print("DLL is empty can't delete")
            return
        if self.head.nref is None:        #####  it shows only one node present in DLL
            if x == self.head.data:         ####  check if data match with the data of 
                self.head  = None
            else:
                print("x is not present in DLL")
                
            return 
        
        if self.head.data ==x:          #### if DLL contains more than 1 node and we want to delete 1 node
            self.head = self.head.nref
            self.head.pref = None
            return
        n = self.head
        while n.nref is not None:        #### for deleting middle nodes
            if x ==n.data:           
                break
            n = n.nref
        if n.nref is not None:
            n.nref.pref = n.pref
            n.pref.nref = n.nref
            
        else:                           #### for deleting last node
            if n.data ==x:
                n.pref.nref =None
            else:
                print("x not in DLL")
                
    
dll = DLL()            
dll.add_begin(4)
dll.add_after(10,4)
dll.delete_by_value(10)
dll.print__LL() 
#dll.print__LL_reverse()
```
