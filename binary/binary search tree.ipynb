```python
class BST:
    def __init__(self,key):
        self.key=key
        self.lchild=None
        self.rchild=None
        
    def insert(self,data):
        if self.key is None:
            self.key = data
            return
    ### if tree is not empty
    ## then we need to check the position of  node that is placing the node 
    ## in the left or right of the root node
   
        if self.key >data :
                if self.lchild:
                    self.lchild.insert(data)        ### insert method is called because it will call recursively
                else:
                    self.lchild = BST(data)
        else:           
                if self.rchild:
                    self.rchild.insert(data)
                else:
                    self.rchild = BST(data)

root=BST(None)        ### now it is an empty tree
root.insert(20)
```

```python
class BST:
    def __init__(self,key):
        self.key=key
        self.lchild=None
        self.rchild=None
        
    def insert(self,data):
        if self.key is None:
            self.key = data
            return
        if self.key==data:  ### when ignoring the duplicate value
            return             
    ### if tree is not empty
    ## then we need to check the position of  node that is placing the node 
    ## in the left or right of the root node
   
        if self.key >data :
                if self.lchild:
                    self.lchild.insert(data)        ### insert method is called because it will call recursively
                else:
                    self.lchild = BST(data)
        else:           
                if self.rchild:
                    self.rchild.insert(data)
                else:
                    self.rchild = BST(data)

    def search(self,data):
        if self.key==data:
            print("Node is found")
            return
        if data<self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node is not present in tree")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node is not present")
                
                
root=BST(10)        ### now it is an empty tree
list1=[20,4,30,4,1,5,6]
for  i in list1:
    root.insert(i)
root.search(6)  
root.search(60)
```

### In-Order

```python
class BST:
    def __init__(self,key):
        self.key=key
        self.lchild=None
        self.rchild=None
        
    def insert(self,data):
        if self.key is None:
            self.key = data
            return
        if self.key==data:  ### when ignoring the duplicate value
            return             
    ### if tree is not empty
    ## then we need to check the position of  node that is placing the node 
    ## in the left or right of the root node
   
        if self.key >data :
                if self.lchild:
                    self.lchild.insert(data)        ### insert method is called because it will call recursively
                else:
                    self.lchild = BST(data)
        else:           
                if self.rchild:
                    self.rchild.insert(data)
                else:
                    self.rchild = BST(data)

    def search(self,data):
        if self.key==data:
            print("Node is found")
            return
        if data<self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node is not present in tree")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node is not present")
                
    def preorder(self):
        print(self.key)
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()
    
    def inorder(self):
        
        if self.lchild:
            self.lchild.inorder()
        print(self.key,end=" ")
        if self.rchild:
            self.rchild.inorder()
                
root=BST(10)        ### now it is an empty tree
list1=[20,4,30,4,1,5,6]
for  i in list1:
    root.insert(i)
print("Pre-Order")    
root.preorder()
print("In-Order")
root.inorder()
```

```python
class BST:
    def __init__(self,key):
        self.key=key
        self.lchild=None
        self.rchild=None
        
    def insert(self,data):
        if self.key is None:
            self.key = data
            return
        if self.key==data:  ### when ignoring the duplicate value
            return             
    ### if tree is not empty
    ## then we need to check the position of  node that is placing the node 
    ## in the left or right of the root node
   
        if self.key >data :
                if self.lchild:
                    self.lchild.insert(data)        ### insert method is called because it will call recursively
                else:
                    self.lchild = BST(data)
        else:           
                if self.rchild:
                    self.rchild.insert(data)
                else:
                    self.rchild = BST(data)

    def search(self,data):
        if self.key==data:
            print("Node is found")
            return
        if data<self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node is not present in tree")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node is not present")
                
    def preorder(self):
        print(self.key)
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()
    
    def inorder(self):
        
        if self.lchild:
            self.lchild.inorder()
        print(self.key,end=" ")
        if self.rchild:
            self.rchild.inorder()
    
    def  postorder(self):
        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key)    
    
    
    def delete(self,data):
        if self.key is None:
            print("Tree is empty")
            return   
            
        if  data < self.key: 
            if self.lchild:
                 self.lchild=self.lchild.delete(data)
            else:
                print("given node is not presenent in hte tree")
            
        elif data > self.key:
            if self.rchild:
                self.rchild=self.rchild.delete(data)
            else:
                print("given node is not presenent in hte tree")
       
        else:               ### the condition written in this will delete the node which contain 0 or 1 child nodes
            if self.lchild is None:
                temp=self.rchild  ### if lchild is none then return rchild
                self=None
                return temp
            if self.rchild is None:
                temp = self.lchild  ### if rchild is none then return lchild
                self=None
                return temp
            
            node =self.rchild                            ####  because we want to fnd the smallest node in the right subtree
            while node.lchild:       ## for 2 child                ### after this node will point t the node which contains the smallest key
                node = node.lchild
            self.key =  node.key
            self.rchild=self.rchild.delete(node.key)
        return self  
root=BST(10)        ### now it is an empty tree
list1=[20,4,30,4,1,5,6]
for  i in list1:
    root.insert(i)


root.delete(600)
print("after deleting")
root.preorder()
root.delete(10)
print("after deleting")
root.preorder()
```
