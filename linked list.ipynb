```python
###  how to check if L.L. is empty 
### when head is none or null then it means L.L. is empty but if head contains some value then it is not
### if head is none then print L.L. is empty
```

```python
class Node:                                       #######   to create node (1st step)
    def __init__(self,data):
        self.data=data
        self.ref=None
node1=Node(10)
print(node1)

class linkedlist:
    def __init__(self):
        self.head=None         ######  means creating empty node ###  head is the start point of linked list  and none is the empty linked list

    def print__LL(self):         ###     condition 1    when linked list is empty   ####  from here to ### 3. line is the way how to traverse through the linked list
        if self.head is None:
            print("L.L. is empty")
        else:
            n=self.head                        
            while n is not None:            
                print(n.data)
                n=n.ref                         ####   3.
LL1=linkedlist()          
LL1.print__LL()
```
